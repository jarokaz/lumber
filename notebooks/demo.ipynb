```python
import matplotlib.image as mpimg
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import json
from PIL import Image
from os.path import join
import utils

%matplotlib inline
%env GOOGLE_APPLICATION_CREDENTIALS=../auth/MLSandbox-auth.json
```

## Load and display test images

```python
images, labels = utils.load_images()
figsize=(10, 7)
fig, axis = plt.subplots(len(images)//3, 3, figsize=figsize)
fig.tight_layout()
for ax, image, label in zip(axis.flat[0:], images, labels):
    ax.set_title(label)
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.imshow(image)

    
```

```python
labels, probs = utils.classify_images(images)

figsize=(10, 7)
fig, axis = plt.subplots(len(images)//3, 3, figsize=figsize)
fig.tight_layout()
for ax, image, label, prob in zip(axis.flat[0:], images, labels, probs):
    ax.set_title(label + ':' + str(prob))
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.imshow(image)
```
