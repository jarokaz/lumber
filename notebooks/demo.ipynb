## Load and display test images

```python
images, labels = utils.load_images(images_dir, test_images1)
figsize=(10, 7)
fig, axis = plt.subplots(len(images)//3, 3, figsize=figsize)
fig.tight_layout()
for ax, image, label in zip(axis.flat[0:], images, labels):
    ax.set_title(label)
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.imshow(image)

```

## Call prediction service and display results

```python
labels, probs = utils.classify_images(images)

figsize=(10, 7)
fig, axis = plt.subplots(len(images)//3, 3, figsize=figsize)
fig.tight_layout()
for ax, image, label, prob in zip(axis.flat[0:], images, labels, probs):
    ax.set_title(label + ':  ' + str(prob))
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.imshow(image)
```
